```python
import tensorflow as tf
import pandas as pd
import numpy as np
keras = tf.keras
from sklearn.utils import shuffle
```

```python
train_data = pd.read_csv('./data/dataAll.csv', index_col=None)
test_featues = pd.read_csv('./data/testAll.csv', index_col=None)

train_data = shuffle(train_data)

train_features = train_data.iloc[:, :-2]  # (3863, 30)
train_labels = train_data.iloc[:, -2:]  # (3863, 2)
```

```python
class MyModel(keras.Model):
    def __init__(self, output_num=2):
        super().__init__(name='my_model')
        self.output_num = output_num
        self.my_layers = [
            tf.layers.Dense(30, activation='relu', kernel_regularizer='l2'),
            tf.layers.Dense(30, activation='relu', kernel_regularizer='l2'),
            tf.layers.Dense(30, activation='relu', kernel_regularizer='l2'),
            tf.layers.Dense(20, activation='relu', kernel_regularizer='l2'),
            tf.layers.Dense(10, activation='relu', kernel_regularizer='l2'),
            tf.layers.Dense(2)
        ]
        
    def call(self, inputs):
        x = self.my_layers[0](inputs)
        for layer in self.my_layers[1:]:
            x = layer(x)
        return x
            
    def compute_output_shape(self, input_shape):
        """Compute the output array shape."""
        shape = tf.TensorShape(input_shape).as_list()
        shape[-1] = self.output_num
        return tf.TensorShape(shape)
        
```

```python
model = MyModel(2)
model.compile(optimizer='adam', loss='mse', metrics=['mse'])
model.fit(train_features.values, train_labels.values, batch_size=32, epochs=40, validation_split=0.2)
```
